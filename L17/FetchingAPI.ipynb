```python
df= pd.DataFrame()
```

```python
for i in range(1,429):
    url= f"https://api.themoviedb.org/3/movie/top_rated?api_key=ac8d19e66344fe5a0910360da49d6a70&language=en-US&page={i}"
    response = requests.get(url)
    data= pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df=pd.concat([df,data],ignore_index=True)
```
